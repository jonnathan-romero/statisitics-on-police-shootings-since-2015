```python
import pandas as pd
import matplotlib.pyplot as plt

pd.options.display.max_rows = 1000
pd.options.display.max_columns  = 1000
```

```python
url = 'https://raw.githubusercontent.com/jonnathan-romero/data-police-shootings/master/fatal-police-shootings-data.csv'
```

```python
df = pd.read_csv(url,index_col='id',parse_dates=['date'])
df['date.year'] = df['date'].map(lambda x: x.year)
df['date.month'] = df['date'].map(lambda x: x.month)
df['date.day'] = df['date'].map(lambda x: x.day)
df['armed_simple'] = ~(df['armed'] == 'unarmed')
```

```python
df.head(10)
```

```python
df.tail(50)
```

```python
df.describe()
```

<h2>How many people died per year?</h2>

```python
(df.groupby(by=['date.year']).describe())[[('date.month','count')]].plot()
```

<h2>Lets break that number by Gender</h2>

```python
(df.groupby(by=['date.year','gender']).describe())[[('date.month','count')]].unstack(1)
```

```python
(df.groupby(by=['date.year','gender']).describe())[[('date.month','count')]].unstack(1).plot()
```

<h2>Lets break that number by Race</h2>

```python
(df.groupby(by=['date.year','race']).describe())[[('date.month','count')]].unstack(1)
```

```python
(df.groupby(by=['date.year','race']).describe())[[('date.month','count')]].unstack(1).plot()
```

<h2>Lets break that number by State</h2>

```python
temp = (df.groupby(by=['date.year','state']).describe())[[('date.month','count')]].unstack(1).T
```

```python
(df.groupby(by=['date.year','state']).describe())[[('date.month','count')]].unstack(1).plot(legend=None)
```

```python
temp[temp.sum(axis=1)>160].T.plot()
```

<h2>Signs of Illness?</h2>

```python
(df.groupby(by=['date.year','signs_of_mental_illness']).describe())[[('date.month','count')]].unstack(1)
```

```python
(df.groupby(by=['date.year','signs_of_mental_illness']).describe())[[('date.month','count')]].unstack(1).plot()
```

<h2>Does Mental Illness affect races the same way?</h2>

```python
(df.groupby(by=['date.year','race']).describe())[[('date.month','count')]].unstack(1)
```

```python
(df.groupby(by=['date.year','signs_of_mental_illness','race']).describe())[[('date.month','count')]].unstack(2).unstack(1)
```

<h3>Seems like white people are more likely to be classified as showing signs of mental illness</h3>

<h2>Another interesting Question is regarding the average age by race?</h2>

```python
(df.groupby(by=['date.year','race']).describe())[[('age','mean')]].unstack(1)['age']['mean'][['B','H','W']].plot()
```

<h3>Hispanic and Blacks tend to be much younger than Whites, but could mental illness be playing a role in this?</h3>

```python
(df.groupby(by=['date.year','signs_of_mental_illness','race']).describe())[[('age','mean')]].unstack(1)['age']['mean'][False].unstack(1)[['B','H','W']].plot()
```

<h3>Mental Illness doesn't seem to play to huge of a issue</h3>

<h2>Now lets looked at unarmed vs. armed?</h2>

```python
(df.groupby(by=['date.year','armed_simple']).describe())[[('date.month','count')]].unstack(1)
```

```python
(df.groupby(by=['date.year','armed_simple']).describe())[[('date.month','count')]].unstack(1).plot()
```

<h3>Majority are armed, but what race gender etc are these people, but one thing to note is that unarmed has been going down</h3>

```python
(df[df['armed_simple']==False].groupby(by=['date.year','race']).describe())[[('date.month','count')]].unstack(1).plot()
```

```python
(df[df['armed_simple']==False].groupby(by=['date.year','gender']).describe())[[('date.month','count')]].unstack(1).plot()
```

<h3>As a percentage more black people are killed unarmed as a percent of total killed in their race?, but black unarmed deaths have actually been going down the most.</h3>
